```python
df_crashes = pd.read_csv("../data/traffic_crashes/traffic_crashes.csv")
```

```python
print(df_crashes.shape)
df_crashes.head()
```

```python
df_crashes.info()
```

```python
df_crashes.isnull().sum()
```

```python
df_crashes = df_crashes.dropna(axis="columns", how="all")
df_crashes.shape
```

```python
df_crashes = df_crashes.dropna(axis="index", thresh=2)
df_crashes.shape
```

```python
df_crashes.report_type.unique()
```

```python
df_crashes = df_crashes.fillna(value={"report_type": "ON SCENE"})
```

```python
df_crashes.report_type.unique()
```

```python
df_vehicles = pd.read_csv("../data/traffic_crashes/traffic_crash_vehicle.csv")
df_vehicles.head()
```

```python
df = df_crashes.merge(df_vehicles, how="left", on="crash_record_id", suffixes=("_left", "_right"))
df.head()
```

```python
df_agg = df.groupby("vehicle_type").agg(({"crash_record_id": "count"})).reset_index()
```

```python
number_of_passengers_cars_involved = df_agg[df_agg["vehicle_type"] == "PASSENGER"].crash_record_id.array[0]
number_of_passengers_cars_involved
```
